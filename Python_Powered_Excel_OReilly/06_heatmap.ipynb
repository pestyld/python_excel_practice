# Creating a heatmap 

```python
import numpy as np
import pandas as pd
```

```python
athletes = pd.read_excel('../data/athletes.xlsx')
athletes.head()
```

## Heatmap

```python
athletes_corr = (
    athletes
    .corr()
    .round(2)
    .replace(1,'')
)

athletes_corr
```

Total rows and columns in correlation matrix.

```python
first_row=1
first_col=1

(last_row, last_col) = athletes_corr.shape

display(first_row, first_col, last_row, last_col)
```

## Create Excel

```python
## Set pandas engine to xlsxwriter
writer = pd.ExcelWriter('../output/13_correlation.xlsx', engine = 'xlsxwriter')

## Convert DF to xlsxwriter object
athletes_corr.to_excel(writer, index = True)

## Create workbook and worksheet
workbook = writer.book
worksheet = writer.sheets['Sheet1']

## Conditional formatting

worksheet.conditional_format(first_row, first_col, last_row, last_col,
                            {'type':'3_color_scale'})


workbook.close()
```
